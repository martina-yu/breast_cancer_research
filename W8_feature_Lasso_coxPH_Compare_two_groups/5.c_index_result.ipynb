```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./c_index_results.xlsx')
# df = pd.read_excel('./c_index_top_n.xlsx')
```

```python
df = df.set_index(df.columns[0])
sns.set_theme(style="white")
plt.figure(figsize=(10, 8)) # 设置图表大小
heatmap = sns.heatmap(df, annot=True, fmt=".4f", cmap="YlGnBu", linewidths=.5, cbar_kws={'label': 'C-index'})
heatmap.set_title('C-index for Four Models', fontdict={'fontsize':16}, pad=12)
plt.show()

plt.savefig('c_index_heatmap.png')
```

```python
data = {
    '40': [0.642],
    '38': [0.653],
    '35': [0.634],
    '30': [0.625],
    '25': [0.703]
}

# Create a DataFrame
df = pd.DataFrame(data, index=['CoxPH'])

# Create a heatmap
plt.figure(figsize=(8, 2)) # Adjust figure size for better display
sns.heatmap(df, annot=True, cmap='YlGnBu', fmt=".3f", linewidths=.5, vmin=0.50,vmax=0.78)

# Set the title
plt.title('Heatmap of CoxPH Values')

# Display the plot
plt.show()

```
